```python
first_df = pd.read_csv('first.csv', header=None)
second_df = pd.read_csv('second.csv', header=None)
```

```python
fr = 0
to = 1000
plt.figure(figsize=(20,20))
for i in range(first_df.shape[1]):
    if i == 0:
        plt.plot(first_df.iloc[fr:to, i] + 5e-5, label='first', color='orange')
        plt.plot(second_df.iloc[fr:to, i] - 5e-5, label='second', color='blue')
    else:
        plt.plot(first_df.iloc[fr:to, i] + 5e-5, color='orange')
        plt.plot(second_df.iloc[fr:to, i] - 5e-5, color='blue')
    plt.legend(loc='best')
    plt.title(f'{i}-th channel')
    
plt.show()
```
